```python
import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42) # You can use any integer value here. For quiz put 42

# Generate 25 random names for the 'nombre' column
nombres = [f'Estudiante{i}' for i in range(10)]

# Generate 4 columns of random grades between 0 and 5 with two decimal places
# np.random.uniform generates floats between low (inclusive) and high (exclusive)
parciales = np.round(np.random.uniform(0, 5, size=(10, 4)), 2);

# Create the DataFrame
df = pd.DataFrame({
    'nombre': nombres,
    'parcial1': parciales[:, 0],
    'parcial2': parciales[:, 1],
    'parcial3': parciales[:, 2],
    'parcial4': parciales[:, 3]
})

df
```

```python
# Set the 'nombre' column as the index immediately after creation
df = df.set_index('nombre')

# Display the DataFrame
df
```

```python
# Calcular el percentil 25 de 'parcial1'
L = [i/10 for i in range(1,11)]
percentile_25 = df.quantile(L)
percentile_25
```

```python
def ubicacion(n):
  nota = df.iloc[n,0]
  percentil = 0
  while percentile_25.iloc[percentil, 0] < nota:
    percentil += 1
  return percentil
```
